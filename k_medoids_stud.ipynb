```python
import pandas as pd

data = pd.read_csv("StudentsPerformance.csv")

data.head()
```

```python
from sklearn_extra.cluster import KMedoids

X = data[['math score','reading score']]

kmedoids = KMedoids(n_clusters=5, random_state=0)
kmedoids.fit(X)

labels = kmedoids.labels_
medoids = kmedoids.cluster_centers_
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(X['math score'], X['reading score'], c=labels, cmap='viridis')

plt.scatter(medoids[:,0], medoids[:,1],
            color='red', marker='X', s=200)

plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.title("Student Clustering using K-Medoids")

plt.show()
```
